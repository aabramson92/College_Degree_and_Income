```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from selenium import webdriver
```

```python
url = 'https://www.dol.gov/whd/state/stateMinWageHis.htm'
response = requests.get(url).text
soup = BeautifulSoup(response)
tables = soup.find_all('table')
wages = pd.read_html(response)
wage = wages[-1]
wage = wage.drop(columns=[1, 2, 4])
wage = wage.drop([0])
wage = wage.drop([53])
wage = wage.drop([54])
wage = wage.drop([55])
wage = wage.rename(index=str, columns={0:"state", 3:'wage2016'})
wage = wage.sort_values(by='wage2016').reset_index()
```

```python
wage['wage2016'] = [x.split('(')[0] for x in wage['wage2016']]
wage['wage2016'] = [x.split('[')[0] for x in wage['wage2016']]
```
